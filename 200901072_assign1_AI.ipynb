```python
 import networkx as nx
import matplotlib.pyplot as plt
from queue import Queue
class path_Recovery:
    def __init__(self,path,cost):
        self.route = path 
        self.cost=cost

class graphF:
    
    def __init__(self):
        self.frontier = Queue()
        self.frontierDfs = []
        self.explored= []
        
    def BFS(self,start,goal,graph):
        self.explored=[]
        pathS= path_Recovery([start],[0])
        
        self.frontier.put((start,pathS))
        
       # node ,path = self.frontier.get()
       # print(pathS.route)
        
        while not ( self.frontier.empty()):
            node,path = self.frontier.get()
        
           # print("ndoe=>",node,path.route)
            if node == goal:
                print("goal Reachead!")
                print("Path Route:",path.route)
                print("Cost :",path.cost," Total:",sum(path.cost))
                return path.route
            if node  not in self.explored and node != goal:
                #print(node)
                self.explored.append(node)
                for x in range (len(graph[node])):
                        cost = graph[node][x][1]
                        path1 = path_Recovery((path.route + [graph[node][x][0]]),[cost]+path.cost)
                        self.frontier.put((graph[node][x][0] , path1))
   
    def DFS(self,start,goal,graph):
    
        pathS= path_Recovery([start],[0])
        self.explored=[]
        
        self.frontierDfs.append((start,pathS))
        
       # node ,path = self.frontier.get()
        #print(self.frontierDfs[0][1].route)
        
        while len(self.frontierDfs) != 0:
            node,path =  self.frontierDfs.pop()
        
           # print("ndoe=>",node,path.route)
            if node == goal:
                print("goal Reachead!")
                print("Path Route:",path.route)
                print("Cost :",path.cost," Total:",sum(path.cost))
                return path.route
            if node  not in self.explored and node != goal:
                #print(node)
                self.explored.append(node)
                for x in range (len(graph[node])):
                        cost = graph[node][x][1]
                        path1 = path_Recovery((path.route + [graph[node][x][0]]),[cost]+path.cost)
                        self.frontierDfs.append((graph[node][x][0] , path1))            

                
            
                
if __name__ == "__main__":
    
    obj = graphF()
    bfs = obj.BFS("Arad","Bucharest",graph)
    dfs =obj.DFS("Arad","Bucharest",graph)
    print("BFS has out-perfomed the DFS")

 
    





```
